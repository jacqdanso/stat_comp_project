Code for cleaning first 10,000 FCC comments. Should turn into a class later in order to do the same for remaining comments.

```python
import numpy as np 
import pandas as pd
import json
import numpy as np
from itertools import chain
```

```python
# open the first of the 10,000 comment packets
data = json.load(open('2017-2018-fcc_comments/set1/ecfs_17-108_0.json'))
```

```python
# delete the first two comments -- they are FCC public notices 
data = data[2:]
# check to see if done correctly
np.size(data)
```

```python
# data is organized in dictionary format
# construct a dataframe containing only relevant columns
fcc_data = pd.DataFrame(data)
fcc_data.drop(['_index', 'attachments', 'authors', 'bureaus', 'confirmation_number', 'date_disseminated',\
              'documents','id_submission','file_number','entity','lawfirms','proceedings',\
              'submissiontype','total_page_count','presented_to','filingstatus'], axis=1)
```

```python
# clean up data for easy indexing 
names = [i[0]['name'] for i in chain(fcc_data['filers'])]
split_names = [i.split() for i in chain(names)]

# use empty string to indicate absence of middle or last names
for i in range(len(split_names)):
    if len(split_names[i]) == 1: # for those with one name
        split_names[i].insert(1,' '); split_names[i].insert(2,' ')
    if len(split_names[i]) == 2: # for those with two names
        split_names[i].insert(2,split_names[i][1]); split_names[i][1] = ' '
        
# ***** Names ***** #       
first_name = [i[0] for i in chain(split_names)]
middle_name = [i[1] for i in chain(split_names)] 
last_name = [i[2] for i in chain(split_names)]  

# ***** Address Info ***** # 
address_line_1= [i['address_line_1'] for i in chain(fcc_data['addressentity'])]
# replace address nans or empty keys with empty string
address_line_2= [i['address_line_2'] if type(i) is dict and 'address_line_2' in i 
                 else ' ' for i in chain(fcc_data['addressentity'])]
international_address=[i['addresstext'] if type(i) is dict and 'addresstext' in i else ' '
                      for i in chain(fcc_data['internationaladdressentity'])]
city= [i['city'] for i in chain(fcc_data['addressentity'])]
state= [i['state'] for i in chain(fcc_data['addressentity'])]
zip5 = [i['zip_code'] for i in chain(fcc_data['addressentity'])]
zip4 = [i['zip_code'] if type(i) is dict and 'zip_code' in i else ' ' for i in chain(fcc_data['addressentity'])]

# ***** Others ***** # 
viewing_status = [i['description'] for i in chain(fcc_data['viewingstatus'])]

# ***** Turn NaNs to strings ***** # 
text_data = [i if type(i) is str else ' ' for i in chain(fcc_data['text_data'])]
text_data_docs = [i[0]['data'] if type(i) is list else ' ' for i in chain(fcc_data['text_data_docs'])]
```

```python
fcc_dict = {'first_name': first_name,
            'middle_name': middle_name,
           'last_name':last_name,
           'email':fcc_data['contact_email'],
           'email_confirmation': fcc_data['emailConfirmation'],
           'address_line_1':address_line_1,
           'address_line_2':address_line_2,
           'city':city,
           'state':state,
           'zip5':zip5,
           'zip4':zip4,
           'international_address':international_address,
           'express_comment':fcc_data['express_comment'],
            'text_data':text_data,
            'text_data_docs':text_data_docs,
           'date_received':fcc_data['date_received'],
           'date_submission':fcc_data['date_submission'],
            'viewing_status':viewing_status}
```

```python
fcc_df = pd.DataFrame(fcc_dict)
```

```python
fcc_df.to_csv('cleaned_ecfs_17-108_0.csv')
```
